```python
# Cell 1: Load Cohort and Analyze Pathways
import sys
sys.path.append('../src')
import pandas as pd
from pathway_analyzer import analyze_oncology_pathways

# Load saved cohort
patients = pd.read_csv('outputs/oncology_patients.csv')
diagnoses = pd.read_csv('outputs/oncology_diagnoses.csv')

print(f"üìä Analyzing pathways for {len(patients):,} patients...")

```

```python
# Cell 2: Run Pathway Analysis
patterns = analyze_oncology_pathways(patients, diagnoses)

print(f"\nüõ§Ô∏è  PATHWAY ANALYSIS RESULTS:")
print(f"Total pathways: {patterns['total_pathways']}")
print(f"Average pathway length: {patterns['pathway_length_stats']['mean']:.1f}")
print(f"Pathway length range: {patterns['pathway_length_stats']['min']}-{patterns['pathway_length_stats']['max']}")
```

```python
# Cell 3: Show Common Patterns
print(f"\nüîÄ MOST COMMON TRANSITIONS:")
for transition, count in patterns['top_transitions'][:5]:
    print(f"  {transition}: {count} patients")

print(f"\nüìà MOST COMMON SEQUENCES:")
for sequence, count in patterns['top_sequences'][:5]:
    print(f"  {sequence}: {count} patients")

print(f"\n‚úÖ Pathway analysis complete!")
print(f"Results saved to outputs/pathway_analysis.json")
```
